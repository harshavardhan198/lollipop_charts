```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('20 Largest Solar Power Plants.xlsx')
df.head()
```

```python
df['NamewithCountry'] = df['Name'] + df['Country']
```

```python
df.columns
```

```python
x  = df['NamewithCountry']
y = df['MWp (Megawatts Peak Capacity)']
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,8))
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.stem(x,y, use_line_collection=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hlines(y =x,xmin = 0, xmax=y, color = 'red' )
plt.plot(y, x, 'o')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hlines(y =x,xmin = 0, xmax=y, color = 'red' )
plt.plot(y, x, 'o', color = 'k')
plt.show()
```
